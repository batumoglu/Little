```python
import pandas as pd
import numpy as np
import datetime
```

```python
data = pd.read_csv('VIP-X030_5.txt')
data.head()
```

```python
data.drop(columns=["VIP'VIP-X030", "5465499000", "B"], inplace=True)
data.columns = ['Datetime', 'Open', 'High','Low','Close','Volume']
```

```python
data['Datetime'] = pd.to_datetime(data['Datetime'], format='%Y%m%d%H%M')
```

```python
data2 = data[(data['Datetime'].dt.year == 2019) & (data['Datetime'].dt.month < 5)]
```

```python
data2.set_index('Datetime', drop=True, inplace=True, verify_integrity=True)
data2.head()
```

```python
agg = data2.resample('15T', kind='period').agg({'Open':'first',
                                'Close':'last',
                                'High':np.max,
                                'Low':np.min,
                                'Volume':np.sum})
agg.head()
```

```python
agg = agg[agg['Close']>0]
```

```python
agg.reset_index(inplace=True)
```

```python
agg['Datetime'] = agg['Datetime'].astype(str)
```

```python
agg.head()
```

```python
agg['Datetime'] = pd.to_datetime(agg['Datetime'], format='%Y-%m-%d %H:%M' )
```

```python
agg['Date'] = agg['Datetime'].dt.date
agg['Time'] = agg['Datetime'].dt.time
agg.head()
```

```python
agg.to_csv('VIOP15min.csv')
```
